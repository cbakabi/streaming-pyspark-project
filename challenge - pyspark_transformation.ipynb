```python
import csv 
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("SpotifySongs").getOrCreate()

df = spark.read.csv("spotify-2023.csv", header=True, inferSchema=True)
```

```python
df_from_csv_with_sql = spark.read.csv("spotify-2023.csv", header=True, inferSchema=True)
df_from_csv_with_sql.createOrReplaceTempView("SpotifySongs2023")
```

```python
result_all = spark.sql("SELECT * FROM SpotifySongs2023 LIMIT 10")
result_all.show()
```

```python
result = spark.sql("SELECT * FROM SpotifySongs2023 WHERE released_year = 2023 ORDER BY streams DESC LIMIT 5")
result.show()
# ORDER BY streams ASC 
```

```python
result = spark.sql("SELECT * FROM SpotifySongs2023 WHERE released_year = '2023' AND in_spotify_charts BETWEEN 1 AND 5 AND in_apple_charts BETWEEN 1 AND 5 and in_deezer_charts BETWEEN 1 AND 5 and in_shazam_charts BETWEEN 1 AND 5 ORDER BY streams ASC LIMIT 5")
result.show()
# ORDER BY streams ASC 
```

```python
result = spark.sql("SELECT track_name, 'artist(s)_name', released_year, streams, in_spotify_charts, in_apple_charts, in_deezer_charts, in_shazam_charts, (in_spotify_charts + in_apple_charts + in_deezer_charts + in_shazam_charts) AS platform_score FROM SpotifySongs2023 WHERE released_year = '2023' AND in_spotify_charts BETWEEN 1 AND 5 AND in_apple_charts BETWEEN 1 AND 5 AND in_deezer_charts BETWEEN 1 AND 5 AND in_shazam_charts BETWEEN 1 AND 5 ORDER BY platform_score ASC, streams DESC LIMIT 5")
result.show()
```
